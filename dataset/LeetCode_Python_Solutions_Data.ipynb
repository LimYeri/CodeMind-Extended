```python
import numpy as np
import pandas as pd
from datasets import load_dataset
```

```python
from datasets import load_dataset
dataset = load_dataset("LimYeri/LeetCode_Python_Solutions_v2", split = "train").to_pandas()
```

```python
def user_message(data_point):
      user_m = f"I don't know how to solve LeetCode Problem {data_point['id']}: {data_point['title']}. Could you provide a Python solution and explain the approach? "
      return user_m
```

```python
# System message
system_message = "You are a kind coding test teacher.\n\n"

# Function to add conversations
def add_conversations(row):
    return [
              { "from": "system",
                "value": system_message },
              { "from": "user",
                "value": user_message(row) },
              { "from": "assistant",
                "value": row['content'] }
            ]

# Apply the function to each row in the DataFrame
dataset['conversations'] = dataset.apply(add_conversations, axis=1)
```
