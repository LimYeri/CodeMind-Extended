###  토큰 세기

```python
import tiktoken

def count_tokens(row):
  prompt = f"""You are an instructor who teaches coding and explains code. When a user enters code, please provide a detailed explanation of the code. Answer only in English.
  Below is the code for LeetCode problem {row['id']}. Please explain the code in detail. Describe which data structures are used in the code and analyze its time and space complexity.
  code :
  {row['content']}

  Below is the content of LeetCode problem {row['id']}. Please refer to it.

  problem content :

  {row['question_content']}
  """

  enc = tiktoken.encoding_for_model("gpt-3.5-turbo")
  tokens = enc.encode(prompt)
  return len(tokens)

# Apply the function to the DataFrame
df['token_count'] = df.apply(count_tokens, axis=1)
```

### 하나씩 실행

```python
def get_openai_response(row):
  prompt = f"""Below is the code for LeetCode problem {row['id']}. Please explain the code in detail. Describe which data structures are used in the code and analyze its time and space complexity.
  code :
  {row['content']}

  Below is the content of LeetCode problem {row['id']}. Please refer to it.

  problem content :
  {row['question_content']}
  """

  try:
    completions = client.chat.completions.create(
        model='gpt-3.5-turbo-0125',  # Determines the quality, speed, and cost.
        temperature=0.5,            # Level of creativity in the response
        messages=[
            {"role": "system", "content": "You are an instructor who teaches coding and explains code. When a user enters code, please provide a detailed explanation of the code. Answer only in English."},
            {"role": "user", "content": prompt},
        ]
    )
    return completions.choices[0].message.content
  except Exception as e:
    print("error")
    return "error"
```

```python
from tqdm import tqdm

for i in tqdm(range(len(df))):
    df.loc[i, 'response'] = get_openai_response(df.iloc[i])
```

### 여러개씩 실행 (병렬 처리로 시간 단축)

```python
from concurrent.futures import ThreadPoolExecutor
from tqdm import tqdm

# Function to apply get_openai_response and save to the DataFrame
def process_row(i):
    response = get_openai_response(df.iloc[i])
    return i, response

# Use ThreadPoolExecutor for parallel processing
with ThreadPoolExecutor(max_workers=10) as executor:
    results = list(tqdm(executor.map(process_row, range(len(df))), total=len(df)))

# Update the DataFrame with responses
for i, response in results:
    df.loc[i, 'response'] = response
```

```python
error_indices = []
for i, response in enumerate(df['response']):
  if response == "error":
    error_indices.append(i)
print(error_indices)
```

```python
len(error_indices)
```

```python
with ThreadPoolExecutor(max_workers=5) as executor:
    retry_results = list(tqdm(executor.map(process_row, error_indices), total=len(error_indices)))

# Update the DataFrame with responses
for i, response in retry_results:
    df.loc[i, 'response'] = response
```

```python
error_indices = []
for i, response in enumerate(df['response']):
  if response == "error":
    error_indices.append(i)
print(error_indices)
```
